```python
import numpy as np
import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\Admin\Desktop\Intership\bank-additional\bank-additional\bank-additional.csv", sep=';')

# Clean the categorical values for later reuse
from sklearn.preprocessing import LabelEncoder
for col in df.select_dtypes(include='object'):
    df[col] = LabelEncoder().fit_transform(df[col])

# Columns to clean outliers from
columns_to_clean = ['age', 'campaign', 'duration']

# Apply IQR method to each column
for col in columns_to_clean:
    Q1 = np.percentile(df[col], 25)
    Q3 = np.percentile(df[col], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter only the rows within bounds
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Result: df is now without outliers in 'age', 'campaign', 'duration'
print("Outliers removed. New shape of data:", df.shape)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\Admin\Desktop\Intership\bank-additional\bank-additional\bank-additional.csv", sep=';')

# Encode categorical columns so correlation can be computed
from sklearn.preprocessing import LabelEncoder

cat_cols = df.select_dtypes(include=['object']).columns
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Now all columns are numeric
# Compute correlation matrix
corr = df.corr()

# Print the full correlation matrix
print("Full Correlation Matrix:")
print(corr)

# Filter correlations with absolute value >= 0.90
high_corr = corr[(abs(corr) >= 0.90) & (abs(corr) < 1.0)]  # exclude perfect self-correlation (1.0)

# Plot heatmap for only high correlations
plt.figure(figsize=(12, 8))
sns.heatmap(high_corr, annot=True, cmap='Set3', linewidths=0.2, mask=high_corr.isnull())
plt.title("Highly Correlated Features (|correlation| â‰¥ 0.90)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder

# Step 1: Load the dataset
df = pd.read_csv(r"C:\Users\Admin\Desktop\Intership\bank-additional\bank-additional\bank-additional.csv", sep=';')

# Step 2: Encode all categorical columns
label_encoders = {}
for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Step 3: Split into features (X) and target (y)
X = df.drop(columns='y')
y = df['y']

# Step 4: Train Decision Tree
dt = DecisionTreeClassifier(max_depth=4, random_state=0)  # adjust max_depth for readability
dt.fit(X, y)

# Step 5: Define class names using label encoder
cn = label_encoders['y'].classes_

# Step 6: Plot the Decision Tree
plt.figure(figsize=(30, 10))
plot_tree(dt, class_names=cn, feature_names=X.columns, filled=True, rounded=True, fontsize=10)
plt.title("Decision Tree on Bank Dataset")
plt.show()
```
